```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample data\\homeprices.csv')   # area and price is called feature 
df
```

```python
plt.scatter(df.area,df.price)  
plt.show()

# NOTE:-  we can apply regression bcz our data is  linear  at 45 degree
```

```python
plt.scatter(df.area,df.price, marker='+')
plt.show()
```

```python
model = LinearRegression()
model.fit(df[['area']], df.price) # train the model
```

```python
model.predict([[3500]])  # testing the model
```
